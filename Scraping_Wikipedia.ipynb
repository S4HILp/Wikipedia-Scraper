```python
from bs4 import BeautifulSoup
import requests
```

```python
searchQuery = input("Enter your search: ")

splitWords = searchQuery.split(" ")
processedQuery = []

for words in splitWords:
  upperCase = words.title()
  processedQuery.append(upperCase)
webQuery = "_".join(processedQuery)

processedQuery
```

```python
url = 'https://en.wikipedia.org/wiki/'+webQuery

header = {"User-Agent":"Chrome"}
wiki = requests.get(url,headers=header).text

soup = BeautifulSoup(wiki,"html5lib")

for soup in soup.find(name="div",class_="mw-body").findAll(["p",'h1']):
  content = soup.text.strip(" ")
  print(content)
```
